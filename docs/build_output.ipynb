```python
from pathlib import Path


def concatenate_markdown_files(input_files, output_file, add_separator=True):
    """
    Concatenate multiple Markdown files into a single file.

    Parameters:
        input_files (list of str or Path): List of .md file paths to combine.
        output_file (str or Path): Path to the output .md file.
        add_separator (bool): Whether to add a separator between files.

    Returns:
        None
    """
    input_files = [Path(f) for f in input_files]
    output_file = Path(output_file)

    with output_file.open("w", encoding="utf-8") as outfile:
        for i, file_path in enumerate(input_files):
            if not file_path.exists():
                raise FileNotFoundError(f"File not found: {file_path}")

            with file_path.open("r", encoding="utf-8") as infile:
                content = infile.read()
                outfile.write(content)

            # Add separator between files (optional)
            if add_separator and i < len(input_files) - 1:
                outfile.write("\n\n---\n\n")  # Markdown horizontal rule
```
